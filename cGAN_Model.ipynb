```python
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from matplotlib import pyplot as plt

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Custom Dataset
class CustomDataset(Dataset):
    def __init__(self, images, labels):
        self.images = images
        self.labels = labels

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        return self.images[idx], self.labels[idx]

# File paths for circles and lines
file_paths = {
    'basketball': "/content/basketball.npy",
    'envelope': "/content/envelope.npy"
}

all_images = []
all_labels = []

# Load images and assign labels based on their shape
for label, path in enumerate(file_paths.values()):
    images = np.load(path)
    images = images.reshape(-1, 28, 28)
    images = (images - 0.5) / 0.5  # Normalization to [-1, 1]
    all_images.append(images)
    all_labels.append(np.full((images.shape[0],), label))  # Assign each shape its correct label

# Combine all images and labels into single arrays
all_images = np.concatenate(all_images, axis=0)
all_labels = np.concatenate(all_labels, axis=0)

# Convert to PyTorch tensors and add channel dimension
all_images = torch.Tensor(all_images).unsqueeze(1)
all_labels = torch.tensor(all_labels, dtype=torch.long)

dataset = CustomDataset(all_images, all_labels)
dataloader = DataLoader(dataset, batch_size=64, shuffle=True, drop_last=True)

# Debug: Verify image-label pairs in DataLoader
print("Verifying data consistency in the DataLoader:")
for i, (imgs, labels) in enumerate(dataloader):
    print(f"Batch {i+1}:")

    # Set up a grid for displaying images
    num_images = min(64, len(imgs))  # Display at most 9 images (3x3 grid)
    fig, axes = plt.subplots(8, 8, figsize=(8, 8))  # 3x3 grid of subplots

    for j in range(num_images):
        ax = axes[j // 8, j % 8]  # Get the correct position in the grid
        ax.imshow(imgs[j].squeeze(0).cpu(), cmap='gray')
        ax.set_title(f"Label: {labels[j].item()}")
        ax.axis('off')  # Hide axis

    # Remove any extra subplots if less than 9 images
    for k in range(num_images, 9):
        axes[k // 3, k % 3].axis('off')

    plt.tight_layout()  # Adjust layout to fit the images
    plt.show()

    if i == 0:  # Stop after the first batch for debugging
        break

class Generator(nn.Module):
    def __init__(self, z_dim=128, num_classes=2):
        super(Generator, self).__init__()
        self.z_dim = z_dim
        self.num_classes = num_classes

        # Embedding layer to convert labels to a latent representation
        self.label_embedding = nn.Embedding(num_classes, num_classes)

        # Define the generator model with additional ConvTranspose2d layers
        self.model = nn.Sequential(
            # First layer: Upsampling from (z_dim + num_classes) x 1 x 1 to 256 x 7 x 7
            nn.ConvTranspose2d(z_dim + num_classes, 256, kernel_size=7, stride=1, padding=0),
            nn.BatchNorm2d(256),
            nn.ReLU(True),

            # Second layer: Upsample to 128 x 14 x 14
            nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(True),

            # Third layer: Upsample to 64 x 28 x 28
            nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(True),

            # Fourth layer: Further refine to final 1 x 28 x 28 output
            nn.ConvTranspose2d(64, 32, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(True),

            # Final layer: Output single channel (grayscale) image with tanh activation for scaling to [-1, 1]
            nn.ConvTranspose2d(32, 1, kernel_size=3, stride=1, padding=1),
            nn.Tanh()
        )

    def forward(self, z, labels):
        labels = self.label_embedding(labels).float()
        z = torch.cat([z, labels], dim=1)
        z = z.view(-1, self.z_dim + self.num_classes, 1, 1)
        return self.model(z)



class Discriminator(nn.Module):
    def __init__(self, num_classes=2):
        super(Discriminator, self).__init__()
        self.num_classes = num_classes
        self.label_embedding = nn.Embedding(num_classes, 128 * 7 * 7)

        self.model = nn.Sequential(
            nn.Conv2d(1, 64, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(64),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Flatten()
        )

        self.fc = nn.Sequential(
            nn.Linear(128 * 7 * 7 + 128 * 7 * 7 , 1),
            nn.Sigmoid()
        )

    def forward(self, img, labels):
        features = self.model(img)
        label_features = self.label_embedding(labels)

        # Strengthen conditioning by adding label features to image features
        combined_features = torch.cat([features, label_features], dim=1)
        return self.fc(combined_features)


# Hyperparameters and model initialization
z_dim = 128
num_classes = 2
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

generator = Generator(z_dim, num_classes).to(device)
discriminator = Discriminator(num_classes).to(device)
optimizer_G = optim.Adam(generator.parameters(), lr=0.0002, betas=(0.5, 0.999))
optimizer_D = optim.Adam(discriminator.parameters(), lr=0.0001, betas=(0.5, 0.999))
adversarial_loss = nn.BCELoss()

g_losses = []
d_losses = []
real_accuracies = []
fake_accuracies = []

def smooth_curve(values, window_size=50):
    smoothed = np.convolve(values, np.ones(window_size)/window_size, mode='valid')
    return smoothed

# Training loop
epochs = 30
for epoch in range(epochs):
    # During the training loop, right before image generation:
    for i, (imgs, labels) in enumerate(dataloader):
        real_imgs = imgs.to(device)
        labels = labels.to(device)

        # Smoothing labels for discriminator
        smooth_real = torch.FloatTensor(imgs.size(0), 1).uniform_(0.9, 1.0).to(device)
        smooth_fake = torch.FloatTensor(imgs.size(0), 1).uniform_(0.0, 0.1).to(device)

        # Discriminator training
        optimizer_D.zero_grad()
        z = torch.randn(imgs.size(0), z_dim, device=device)

        # Use the real labels to generate fake images
        gen_labels = labels  # Use the real labels, not random labels

        # Generate fake images based on random noise and the real labels
        gen_imgs = generator(z, gen_labels)

        # Optional: Visualize the generated images and their corresponding labels
        if i % 10000 == 0:  # Every 100 batches (or adjust as needed)
            fig, axarr = plt.subplots(1, 5, figsize=(15, 3))  # Display 5 images from this batch
            for j in range(5):
                # Detach the tensor from the computation graph, move it to CPU, and convert to numpy
                img = gen_imgs[j].detach().cpu().squeeze(0).numpy()
                axarr[j].imshow(img, cmap='gray')
                axarr[j].set_title(f"Label: {gen_labels[j].item()}")
                axarr[j].axis('off')
            plt.show()

        real_preds = discriminator(real_imgs, labels)
        fake_preds = discriminator(gen_imgs.detach(), gen_labels)

        # Calculate losses and update discriminator and generator (your existing training steps)
        real_loss = adversarial_loss(discriminator(real_imgs, labels), smooth_real)
        fake_loss = adversarial_loss(discriminator(gen_imgs.detach(), gen_labels), smooth_fake)
        d_loss = (real_loss + fake_loss) / 2
        d_loss.backward()
        optimizer_D.step()

        # Calculate discriminator accuracy
        real_acc = (real_preds > 0.5).float().mean().item() * 100
        fake_acc = (fake_preds < 0.5).float().mean().item() * 100

        # Generator training
        optimizer_G.zero_grad()
        valid = torch.ones((imgs.size(0), 1), device=device)
        g_loss = adversarial_loss(discriminator(gen_imgs, gen_labels), valid)
        g_loss.backward()
        optimizer_G.step()

        # Store losses and accuracy
        g_losses.append(g_loss.item())
        d_losses.append(d_loss.item())
        real_accuracies.append(real_acc)
        fake_accuracies.append(fake_acc)

        if i % 500 == 0:
            print(f"Epoch [{epoch}/{epochs}] Batch {i}/{len(dataloader)} Loss D: {d_loss.item():.4f}, Loss G: {g_loss.item():.4f}, "
                  f"Real Acc: {real_acc:.2f}%, Fake Acc: {fake_acc:.2f}%")


    # After each epoch, generate and display images for each class
    with torch.no_grad():
        fig, axarr = plt.subplots(1, num_classes, figsize=(10, 5))
        for class_label in range(num_classes):
            z = torch.randn(1, z_dim, device=device)
            gen_label = torch.tensor([class_label], device=device)
            fake_img = generator(z, gen_label).cpu().view(28, 28)
            axarr[class_label].imshow(fake_img, cmap='gray')
            axarr[class_label].set_title(f'Class {class_label} ({list(file_paths.keys())[class_label]})')
            axarr[class_label].axis('off')
        plt.show()

# After training, plot losses and accuracy
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(g_losses, label="G Loss")
plt.plot(d_losses, label="D Loss")
plt.xlabel("Batch")
plt.ylabel("Loss")
plt.legend()
plt.title("Generator and Discriminator Loss")

plt.subplot(1, 2, 2)
plt.plot(real_accuracies, label="Real Acc")
plt.plot(fake_accuracies, label="Fake Acc")
plt.xlabel("Batch")
plt.ylabel("Accuracy (%)")
plt.legend()
plt.title("Discriminator Accuracy")
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))

# Loss Plot
plt.subplot(1, 2, 1)
plt.plot(smooth_curve(g_losses, window_size=50), label="G Loss (smoothed)")
plt.plot(smooth_curve(d_losses, window_size=50), label="D Loss (smoothed)")
plt.xlabel("Batch")
plt.ylabel("Loss")
plt.legend()
plt.title("Generator and Discriminator Loss")

# Accuracy Plot
plt.subplot(1, 2, 2)
plt.plot(smooth_curve(real_accuracies, window_size=50), label="Real Acc (smoothed)")
plt.plot(smooth_curve(fake_accuracies, window_size=50), label="Fake Acc (smoothed)")
plt.xlabel("Batch")
plt.ylabel("Accuracy (%)")
plt.legend()
plt.title("Discriminator Accuracy (Smoothed)")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import torch
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from scipy.stats import multivariate_normal
from PIL import Image
import matplotlib.pyplot as plt

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# UnifiedImagesDataset class (reuse from FID calculation)
class UnifiedImagesDataset(Dataset):
    def __init__(self, image_file, normalize=True):
        self.images = np.load(image_file)  # Load .npy file
        self.images = self.images.squeeze()  # Remove singleton dimension

        # Normalize images to [0, 1]
        if self.images.max() > 1:  # Likely in range [0, 255]
            print(f"Rescaling images in {image_file} from [0, 255] to [0, 1]")
            self.images = self.images / 255.0
        elif self.images.min() < 0:  # Already in range [-1, 1]
            print(f"Rescaling images in {image_file} from [-1, 1] to [0, 1]")
            self.images = (self.images + 1) / 2.0

        self.transform = transforms.Compose([
            transforms.Lambda(lambda x: Image.fromarray((x * 255).astype(np.uint8))),  # Convert to 0–255
            transforms.Lambda(lambda img: img.convert("RGB")),  # Convert grayscale to RGB
            transforms.Resize((299, 299), interpolation=Image.NEAREST),  # Resize for InceptionV3
            transforms.ToTensor(),  # Convert to tensor
            transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])  # Normalize to [-1, 1]
        ])

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        image = self.images[idx]
        if image.ndim == 1:  # Flattened image
            side_length = int(np.sqrt(image.shape[0]))
            image = image.reshape((side_length, side_length))
        transformed_image = self.transform(image)
        return transformed_image, image  # Return both transformed and original

# Function to extract InceptionV3 features
def extract_features(image_file, batch_size=32, device='cuda'):
    # Load pre-trained InceptionV3 model
    inception_model = models.inception_v3(pretrained=True, transform_input=False).to(device)
    inception_model.fc = torch.nn.Identity()  # Replace the final layer with Identity
    inception_model.eval()

    dataset = UnifiedImagesDataset(image_file)
    dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=False)

    features = []
    with torch.no_grad():
        for images, _ in dataloader:
            images = images.to(device)
            feat = inception_model(images)
            features.append(feat.cpu().numpy())

    return np.concatenate(features, axis=0)

# Function for anomaly detection
def detect_anomalies(training_file, target_file, threshold=0.01, batch_size=32):
    print("Extracting features for training dataset...")
    training_features = extract_features(training_file, batch_size=batch_size, device=device)
    training_mean = np.mean(training_features, axis=0)
    training_cov = np.cov(training_features, rowvar=False)

    # Multivariate normal distribution using training data statistics
    mvn = multivariate_normal(mean=training_mean, cov=training_cov)

    print("Extracting features for target dataset...")
    target_features = extract_features(target_file, batch_size=batch_size, device=device)

    # Compute likelihood scores
    scores = mvn.logpdf(target_features)
    anomalies = scores < np.log(threshold)

    return scores, anomalies

# Function to visualize anomalies
def visualize_anomalies(target_file, anomalies):
    dataset = UnifiedImagesDataset(target_file)
    fig, axes = plt.subplots(1, 5, figsize=(15, 5))
    idx = 0

    for i in range(len(anomalies)):
        if anomalies[i]:
            _, image = dataset[i]
            axes[idx].imshow(image, cmap='gray')
            axes[idx].set_title(f"Anomaly {idx+1}")
            axes[idx].axis('off')
            idx += 1
            if idx == 5:  # Display 5 anomalies
                break

    plt.tight_layout()
    plt.show()

# Run anomaly detection for basketball and envelope datasets
basketball_training = 'basketball.npy'  # Training dataset for basketball
basketball_target = 'generated_images_basketball.npy'  # Target dataset for basketball

print("\nDetecting anomalies in Basketball dataset...")
scores_basketball, anomalies_basketball = detect_anomalies(basketball_training, basketball_target)
print(f"Detected {np.sum(anomalies_basketball)} anomalies in Basketball dataset.")
visualize_anomalies(basketball_target, anomalies_basketball)

envelope_training = 'envelope.npy'  # Training dataset for envelope
envelope_target = 'generated_images_envelope.npy'  # Target dataset for envelope

print("\nDetecting anomalies in Envelope dataset...")
scores_envelope, anomalies_envelope = detect_anomalies(envelope_training, envelope_target)
print(f"Detected {np.sum(anomalies_envelope)} anomalies in Envelope dataset.")
visualize_anomalies(envelope_target, anomalies_envelope)
```
